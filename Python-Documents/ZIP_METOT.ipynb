```python
Python zip() method returns a zip object, which is an iterator of strings, tuples,lists etc.  where the first item in each passed iterator is paired together, and then the second item in each passed iterator is paired together.

Example :
 
a = [11, 21, 19]
print(zip(a)) # be careful we zipped <list a> with nothing, but no prob for zip ,will generate a series 		of 1-item tuples
print(type(zip(a)))
data = list(zip(a))
print(data)

Example :

val1 = ("Kislay", "Vamsi", "Joson")
val2 = ("Chennai", "Bangalore", "Pune")
res = zip(val1, val2)
print(tuple(res))

Example :

val1 = [56, 57, 58]
val2 = ['a', 'b', 'c']
val3 = [1.0, 2.0, 3.0]
zipped = zip(val1, val2, val3)
print(list(zipped)

Example :

k = ['F_Name', 'L_Name', 'Age'] # list
v = ('Kislay', 'Kumar', '35') # tuple
res_dict = dict(zip(k, v)) # no prob for zip
print(res_dict)

Example :

f_name = ['kislay', 'Kunal', 'Mohit']
age = [24, 50, 18]
for f_name, age in zip(f_name, age):
    print(f_name, age)

Example :

dictA = {
    'name': 'Homer',
    'age': 50,
    'type': 'fat'
}

dictB = {
    'name': 'Bart',
    'age': 10,
    'type': 'medium'
}

for (key1, val1), (key2, val2) in zip(dictA.items(), dictB.items()):
    print(key1, '->', val1)
    print(key2, '->', val2)

Example :

dataStr = ['z', 'y', 'w', 'x']
dataInt = [19, 21, 18, 46]

zippyData = zip(dataInt, dataStr)
data = list(zippyData)
print(data)
print(sorted(data))


Example :

# initializing list of players.
players = [ "Sachin", "Sehwag", "Gambhir", "Dravid", "Raina" ]
  
# initializing their scores
scores = [100, 15, 17, 28, 43 ]
  
# printing players and scores.
for pl, sc in zip(players, scores):
    print ("Player :  %s     Score : %d" %(pl, sc))


unzipping :

We can also extract the data from the Python zip function. If we want to remove the zip, we have to use the same zip()function. But we have to add an asterisk(*) in front of that list you get from the zipped variable.

Example :

val1 = ['a', 'b', 'c']
val2 = [1, 2, 3]
res = zip(val1, val2)
res_list = list(res)
print(res_list)
v1, v2 =  zip(*res_list)
print('v1 =', v1)
print('v2 =', v2)

Example :

pairs = [(1, 'a'), (2, 'b'), (3, 'c'), (4, 'd')]
numbers, letters = zip(*pairs)
print(numbers)
print(letters)


```
